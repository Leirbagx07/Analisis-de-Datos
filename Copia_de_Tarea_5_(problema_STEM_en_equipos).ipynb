**CONTAMINACIÓN DEL AIRE**


La contaminación del aire es la presencia de sustancias químicas, físicas o biológicas en el ambiente que alteran sus características naturales. Esta puede darse tanto en interiores como exteriores, siendo causada por dispositivos de combustión domésticos, vehículos motorizados, instalaciones industriales e incendios forestales. La contaminación del aire es una preocupación de salud pública debido a que incluye partículas como el material particulado, monóxido de carbono, ozono, dióxido de nitrógeno y dióxido de azufre. Tanto la contaminación del aire exterior como interior son causantes de enfermedades respiratorias y otras afecciones, contribuyendo significativamente a la morbilidad y mortalidad.

En el conjunto de datos analizado, se incluye información geolocalizada de los siguientes contaminantes:

1. Dióxido de Nitrógeno [NO2]: Es un gas que se forma principalmente por emisiones de vehículos y plantas de energía. La exposición a NO2 puede agravar enfermedades respiratorias y a largo plazo contribuir al desarrollo de asma e infecciones respiratorias. Los niños, ancianos y personas con asma son especialmente vulnerables.

2. Ozono [O3]: A nivel del suelo, el ozono se produce por reacciones químicas entre óxidos de nitrógeno y compuestos orgánicos volátiles. A diferencia del ozono que se encuentra en la atmósfera superior y que nos protege de los rayos ultravioleta, el ozono terrestre puede causar problemas de salud como dolor de pecho, tos, irritación de garganta e inflamación de las vías respiratorias, además de reducir la función pulmonar y empeorar enfermedades como el bronquitis, enfisema y asma.

3. Monóxido de Carbono [CO]: Este gas incoloro e inodoro es emitido principalmente por vehículos y maquinaria que queman combustibles fósiles. La inhalación de CO puede reducir la capacidad de transporte de oxígeno en la sangre a órganos vitales, causando síntomas como mareos y confusión, y a niveles muy altos, puede ser letal.

4. Material Particulado [PM2.5 y PM10]: Estas pequeñas partículas sólidas o líquidas pueden ingresar al aire y, al ser inhaladas, causar serios problemas cardíacos y pulmonares. El PM2.5 y PM10 se refieren al tamaño de estas partículas. La exposición a estas partículas ha sido clasificada como carcinógeno grupo 1 por la Agencia Internacional de Investigación sobre el Cáncer (IARC), indicando un conocido riesgo de cáncer.

El análisis de estos datos es crucial para entender la distribución y concentración de contaminantes, lo cual es fundamental para desarrollar estrategias de mitigación y políticas de salud pública para proteger a la población.

Además, cabe hablar sobre qué es el AQI antes de comenzar con el proyecto programado:

El **Índice de Calidad del Aire** (AQI, por sus siglas en inglés) es una medida que se utiliza para informar a la población qué tan limpio o contaminado está el aire y cuáles podrían ser los posibles efectos para la salud asociados a la calidad del aire en un momento determinado. El AQI considera varios contaminantes como partículas en suspensión (PM10 y PM2.5), ozono (O3), dióxido de nitrógeno (NO2), monóxido de carbono (CO) y dióxido de azufre (SO2).

El valor del AQI varía entre 0 y 500, donde valores menores representan mejor calidad del aire y valores más altos representan peor calidad del aire. A continuación, se presenta una tabla generalizada con las categorías de AQI:



*  0-50: Buena. La calidad del aire se considera satisfactoria y la contaminación del aire plantea poco o ningún riesgo.
*  51-100: Moderada. La calidad del aire es aceptable; sin embargo, para algunos contaminantes puede haber un problema moderado de salud para un número muy pequeño de personas.
* 101-150: Dañina para grupos sensibles. Los miembros de grupos sensibles pueden experimentar efectos en la salud. Es menos probable que el público en general se vea afectado.
* 151-200: Dañina. Todo el mundo puede comenzar a experimentar problemas de salud; los miembros de grupos sensibles pueden experimentar efectos más serios en la salud.
* 201-300: Muy dañina. Advertencias de salud de emergencia para toda la población.
* 301-500: Peligrosa. Alerta de salud: todos pueden experimentar efectos más graves en la salud.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/global air pollution dataset.csv')
```

```python
df=pd.read_csv('/global air pollution dataset.csv')
data=pd.DataFrame(df)
print(data)
```

```python
def explorar_datos(df):
    print(df.head())
    print(df.describe())
    print(df.info())

explorar_datos(df)
```

```python
# Calculamos las estadísticas descriptivas para cada contaminante
def calcular_estadisticas(dataframe):
    statistics = dataframe.describe()
    return statistics

stats = calcular_estadisticas(df[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']])
print(stats)
```

```python
# Histogramas para cada contaminante
def grafico_histogramas(dataframe):
    dataframe[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].hist(bins=30, figsize=(20, 15))
    plt.suptitle('Distribución de los Valores AQI para Diferentes Contaminantes')
    plt.show()

grafico_histogramas(df)
```

```python
# Boxplots-Diagrama de caja y bigotes para cada contaminante
def grafico_bigotes(dataframe):
    plt.figure(figsize=(10, 8))
    sns.boxplot(data=dataframe[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']])
    plt.title('Diagrama de Cajas y Bigotes de los Valores AQI para Diferentes Contaminantes')
    plt.show()

grafico_bigotes(df)
```

```python
# Gráfico de barras de promedios
def grafico_promedio_aqi(dataframe):
    means = dataframe[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].mean()
    means.plot(kind='bar', figsize=(10, 6))
    plt.title('Promedio de Valores AQI por Contaminante')
    plt.ylabel('Promedio AQI')
    plt.show()

grafico_promedio_aqi(df)
```

```python
# Función para obtener las ciudades con mejor y peor calidad de aire promedio
def rank_ciudades_por_aqi(dataframe):
    city_ranking = dataframe.groupby('City')['AQI Value'].mean().sort_values()
    best_cities = city_ranking.head(5)
    worst_cities = city_ranking.tail(5)
    return best_cities, worst_cities

best_cities, worst_cities = rank_ciudades_por_aqi(df)
print("Ciudades con mejor calidad de aire:", best_cities)
print("Ciudades con peor calidad de aire:", worst_cities)
```

**ANÁLISIS DE GRÁFICOS:**
1. **Histogramas de Distribución:**
 Los histogramas muestran la distribución de los valores AQI para cada contaminante. Es evidente que para el monóxido de carbono (CO) y el dióxido de nitrógeno (NO2), la mayoría de las observaciones se agrupan en el extremo inferior de la escala AQI, indicando en general niveles bajos. En cambio, para el ozono (O3) y las partículas finas (PM2.5), hay una distribución más amplia de valores, con una notable cantidad de observaciones en el rango medio y alto de AQI.

2. **Diagrama de Cajas y Bigotes**
Los diagramas de caja muestran la dispersión y los valores atípicos para cada contaminante. Los valores atípicos en el PM2.5 son notables, lo que indica episodios de muy mala calidad del aire relacionados con este contaminante. La mediana y los cuartiles son bastante bajos para CO y NO2, lo que sugiere que los valores extremadamente altos son menos comunes.

3. **Promedios de AQI por Contaminante:**
 Este gráfico de barras resalta que el PM2.5 tiene el promedio más alto de AQI en comparación con otros contaminantes, lo que sugiere que podría ser la mayor preocupación en términos de contaminación del aire en general.
