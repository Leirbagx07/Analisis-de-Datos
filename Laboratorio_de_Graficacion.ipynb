```python
import pandas as pd
import matplotlib.pyplot as plt #Libreria para trabajar graficacion
```

```python
datos=pd.read_csv("/content/casasboston.csv")
```

```python
print(datos)
```

```python
#df=datos[["RM", "CRIM","MEDV","TOWN", "CHAS"]] #esto si quisiera extraer solo ciertas columnas del DataFrame
```

```python
df=datos.rename(columns={ #Esto me permite cambiar el nombre de las variables
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(color='#FF5733')  #Histograma
plt.title('Histograma de Cantidad de Habitaciones',fontsize=14,fontweight='bold')
plt.xlabel('Cantidad de Habitaciones',fontsize=14,fontweight='bold')
plt.ylabel('Poblacion',fontsize=14,fontweight='bold')
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN", alpha=0.5)
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh() #Ensea los primeros 10
plt.show()
```
